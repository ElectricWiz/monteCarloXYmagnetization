```julia
function boltzProb(pos1, pos2, KT)
    deltaE = 1/(norm(pos2, pos1))^12 - 1/(norm(pos2, pos1))^6
    
    if deltaE == 0 
        return 0
    else 
        return min(1, exp(-deltaE/KT))
    end
end
```

```julia
using Plots, LaTeXStrings, LinearAlgebra

function metropolisDiscosSuavecitos(posiciones, r, δt, KT)
    traslaciones = [[-1+i,-1+j] for i in 0:2 for j in 0:2]
    i = rand(1:length(posiciones))
    θ = 2π*rand()
    v = [cos(θ), sin(θ)]
    posiciones2 = deepcopy(posiciones)
    posiciones2[i] .+= v*δt
    posiciones2[i] = mod.(posiciones2[i], 1)
    q = rand()
    
    for j in (1:(i-1))∪((i+1):length(posiciones2))
        test = true
        for k in 1:9
            if norm(posiciones2[i]+traslaciones[k]-posiciones2[j]) > r 
                test *= true
            else
                if boltzProb(posiciones2[i] + traslaciones[k], posiciones2[j], KT) > q
                    test *= true
                else
                    test *= false
                end
                
            end
            
        end
        
        if !(test)
            return posiciones
        end
        
    end
    
    return posiciones2
end
```
